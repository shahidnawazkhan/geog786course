```python
#Now the next step is extract values for these cells from the original image from each band
water= clipped_img[:, 75, 20]
vegetation= clipped_img[:, 10, 3]
print("value of water in all selected bands i.e. 2-7: ", water)
print("value of vegetation in all selected bands i.e. 2-7: ", vegetation)
```

```python
#Method 1 step 1: DN(Qcal)conversion to TOA Radiance(Lλ)
#declaring variables
# l=m+a where l is TOA spectral radiance, m is band specific multiplicative rescaling factor and a is band specific additive rescaling factor


radiance_mult_fac= [0.00033420,0.00033420,0.00033420]
additive_resc_fac= [0.1,0.1,0.1]

l_water = water*radiance_mult_fac+additive_resc_fac
l_veg = vegetation*radiance_mult_fac+additive_resc_fac


#above two variables stores TOA spectral randiance for all desired bands for two pixels i.e. water and vegetation as shown below
print("ignore 3rd value in both arrays as it is from level 2")
print("TOA Spectral radiance values for Water", l_water)
print("TOA Spectral radiance values for Vegetation", l_veg)
```

```python
#Now to caluclate At sensor brightness temperature we will use planks law
# T= (K2)/ln(1+K1/l) where l is at sensor radiance
k1=[774.88, 480.88, 0.1]
k2=[1321.07, 1201.14,0.1]



water_btemp_k= k2/(np.log(1+k1/l_water))
vege_btemp_k= k2 /(np.log(1+k1/l_veg))

water_btemp_c=water_btemp_k-273.15
vege_btemp_c= vege_btemp_k-273.15

#Now that we only need band 10 temperatures we will make new array and pu

xaxis= ['Water', 'Vegetation']
b_temp_k= [water_btemp_k[0], vege_btemp_k[0]]
b_temp_c= [water_btemp_c[0], vege_btemp_c[0]]

print(water_b_temp)
print(vege_b_temp)

fig, (ax1) = plt.subplots(1,1, figsize=(8,6)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(xaxis, b_temp_k, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Temperature (K)')
ax1.plot(xaxis, b_temp_c, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Temperature (C)')
ax1.legend()
ax1.set(xlabel='Brightness Temperature', ylabel='Value',
title='Brightness Temperature of Selected Pixels')
ax1.grid()
ax1.set_ylim([0, 400])

fig.savefig("A4task1.png", dpi=600)
plt.show()
```

# Surface Temperature Calculation from Level 2 Data

```python
#below values are raw and we need to convert it using the constants provided by Landsat to actual Surface temperatures
waterl2= water[2]
vegel2= vegetation[2]
print("Values in Level 2 data for Water and Vegetation respectively are:")
print(waterl2)
print(vegel2)
```

```python
#using the linear equation to convert these values to surface temperature are here

surf_temp_water= waterl2*0.00341802+149
surf_temp_vege= vegel2*0.00341802+149

print(surf_temp_water)
print(surf_temp_vege)
```

```python
surf_temp_k= [surf_temp_water, surf_temp_vege]
surf_temp_c= [surf_temp_water-273.15, surf_temp_vege-273.15]
print(surf_temp_c)



fig, (ax1) = plt.subplots(1,1, figsize=(8,6)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(xaxis, surf_temp_k, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Temperature (K)')
ax1.plot(xaxis, surf_temp_c, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Temperature (C)')
ax1.legend()
ax1.set(xlabel='Surface Temperature', ylabel='Value',
title='Surface Temperature of Selected Pixels')
ax1.grid()
ax1.set_ylim([0, 400])

fig.savefig("A4task2.png", dpi=600)
plt.show()
```
