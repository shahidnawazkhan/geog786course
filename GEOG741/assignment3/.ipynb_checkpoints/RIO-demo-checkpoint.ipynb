```python
import rasterio  
import matplotlib
%matplotlib inline 
import os
import numpy as np
from rasterio.plot import show, show_hist # some specific rasterio functions we'll need
import matplotlib.pyplot as plt
import folium
print("rasterio's version is: " + rasterio.__version__)
```

```python
#image name m_4409625_sw_14_060_20200831.TIF
# filepath to our image
img_fp = 'highresIm\\m_4409625_sw_14_060_20200831.tif'
# Open a geospatial dataset
dataset = rasterio.open(img_fp)
print(dataset)
```

```python
# what is the name of this image
img_name = dataset.name
print('Image filename: {n}\n'.format(n=img_name))

# How many bands does this image have?
num_bands = dataset.count
print('Number of bands in image: {n}\n'.format(n=num_bands))

# How many rows and columns?
rows, cols = dataset.shape
print('Image size is: {r} rows x {c} columns\n'.format(r=rows, c=cols))

# Does the raster have a description or metadata?
desc = dataset.descriptions
metadata = dataset.meta

print('Raster description: {desc}\n'.format(desc=desc))

# What driver was used to open the raster?
driver = dataset.driver
print('Raster driver: {d}\n'.format(d=driver))

# What is the raster's projection?
proj = dataset.crs
print('Image projection:')
print(proj, '\n')
# What is the raster's "geo-transform"
gt = dataset.transform

print('Image geo-transform:\n{gt}\n'.format(gt=gt))
print('All raster metadata:')
print(metadata)
print('\n')
```

```python
full_img = dataset.read()
print(full_img.shape) # bands, rows, cols

# Open the fourth band in our image - NIR here
nir = dataset.read(4)
print(nir.shape) # check out the dimensions of the image
```

```python
datatype = dataset.dtypes
print('Band datatypes: {dt}'.format(dt=datatype))

# How about some band statistics?
band_mean = np.mean(nir)
band_min = np.amin(nir)
band_max = np.amax(nir)
band_stddev = np.std(nir)
print('Band range: {minimum} - {maximum}'.format(maximum=band_max,
                                                 minimum=band_min))
print('Band mean, stddev: {m}, {s}\n'.format(m=band_mean, s=band_stddev))
```

```python
fig, axr = plt.subplots(1, figsize=(30,12))
#show(nir, transform=dataset.transform, cmap='Blues')

show(nir, ax=axr, cmap='Blues', title='NIR')
```

```python
# open these files and add all bands to an array
arrs = []
for band in range(dataset.count):
    arrs.append(dataset.read(band+1))
# convert the list to a numpy array
aerial_img = np.array(arrs, dtype=arrs[0].dtype)
# let's check the shape of this array
aerial_img.shape
```

```python
clipped_img = aerial_img[:, 0:1400:, 0:2150]
clipped_img.shape
```

```python
fig, (axr, axg, axb, axnir) = plt.subplots(1,4, figsize=(30,12))
show((dataset, 1), ax=axr, cmap='Reds', title='red channel')
show((dataset, 2), ax=axg, cmap='Greens', title='green channel')
show((dataset, 3), ax=axb, cmap='Blues', title='blue channel')
show((dataset, 4), ax=axnir, cmap='Blues', title='NIR channel')
#show(clipped_img[:, :, :], ax=axnir, cmap='Blues', title='NIR channel')
```

```python
rasterio.plot.show_hist(clipped_img, bins=50, histtype='stepfilled', lw=0.0, stacked=False, alpha=0.3)
```

```python
np.seterr(divide='ignore', invalid='ignore')

bandNIR = clipped_img[3] # fourth band
bandRed = clipped_img[2] # second band

# note that in python division of integers leads to integers so we need to specify floats in order to get floats
ndvi = (bandNIR.astype(float)-bandRed.astype(float))/(bandNIR.astype(float)+bandRed.astype(float))
```

```python
print('\nMax NDVI: {m}'.format(m=ndvi.max()))
print('Mean NDVI: {m}'.format(m=ndvi.mean()))
print('Median NDVI: {m}'.format(m=np.median(ndvi)))
print('Min NDVI: {m}'.format(m=ndvi.min()))
```

```python
plt.imshow(ndvi, cmap="RdYlGn")
plt.colorbar()
plt.show()
```

```python
from rasterio.plot import reshape_as_raster, reshape_as_image

# Extract reference to SWIR1, NIR, and Red bands
index = np.array([3, 2, 1])
colors = sentinel_img[index, :, :].astype(np.float64)

max_val = 5000
min_val = 0

# Enforce maximum and minimum values
colors[colors[:, :, :] > max_val] = max_val
colors[colors[:, :, :] < min_val] = min_val

for b in range(colors.shape[0]):
    colors[b, :, :] = colors[b, :, :] * 1 / (max_val - min_val)

# rasters are in the format [bands, rows, cols] whereas images are typically [rows, cols, bands]
# and so our array needs to be reshaped
print(colors.shape)
colors_reshaped = reshape_as_image(colors)
print(colors_reshaped.shape)
```

```python
np.seterr(divide='ignore', invalid='ignore')
bandNIR = sentinel_img[3, :, :]
bandRed = sentinel_img[2, :, :]

ndvi = (bandNIR.astype(float)-bandRed.astype(float))/(bandNIR.astype(float)+bandRed.astype(float))
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5)) 

# Show the color image
axs[0].imshow(colors_reshaped)
axs[0].set_title('Color Image')

# Show NDVI
axs[1].imshow(ndvi, cmap='RdYlGn')
axs[1].set_title('NDVI')
```
