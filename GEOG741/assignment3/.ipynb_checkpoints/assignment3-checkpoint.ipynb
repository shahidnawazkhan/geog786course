```python
#This is the first assignment of Quantitative Remote Sensing for Terrestrial Monitoring (GEOG/GSE-741-S01)
#Import necessary libraries for processing
import os 
import matplotlib as mtp
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt
print("Libraries Imported Successfully")

months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"])
solar_azimuth= np.array([155.63,150.45, 145.43, 138.18, 126.8, 116.21, 118.4, 131.81, 147, 157.18, 161.35, 160.1])
solar_elevation= np.array([30.92,38.21,48.6, 59.92, 67.44, 69.48,  67.29,62.21, 54.14, 44.41, 35.09, 29.95])
solar_zenith= 90-solar_elevation


#print(solar_azimuth)
#print(solar_zenith)

solar_azimuth_rad= solar_azimuth*(np.pi/180)
solar_zenith_rad= solar_zenith*(np.pi/180)

solar_azimuth_rad_squared= np.square(solar_azimuth_rad)
solar_zenith_rad_squared= np.square(solar_zenith_rad)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,6)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(months, solar_azimuth_rad, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Solar Azimuth s')
ax1.plot(months, solar_zenith_rad, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Solar Zenith s')
ax1.legend()
ax1.set(xlabel='Mid-month i.e 15th of each month', ylabel='Value',
title='Solar Zenith and Solar Azimuth in Radians')
ax1.grid()
ax1.set_ylim([0, 3.5])

ax2.plot(months, solar_azimuth, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Solar Azimuth s')
ax2.plot(months, solar_zenith, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Solar Zenith s')
ax2.set(xlabel='Mid-month i.e 15th of each month', ylabel='Value',
title='Solar Zenith and Solar Azimuth in Degrees')
ax2.grid()
ax2.set_ylim([0, 200])
ax2.legend()



fig.savefig("A3task2.png", dpi=600)
plt.show()
```

# G-1 Conditions Calculations for Red and NIR

```python
equation_1= " = a(v^2+s^2)+b(v^2s^2)+cvsCos(v-s)+d"

#Assuming Condition G-1 i.e. Landsat Swath Center where v and v both are zero

red_values = [-0.0606, 0.0755,0.0284, 0.162]
nir_values = [-0.204, 0.296,0.125, 0.454]
#for G-1 Condition v and v both are zero so the so equation 1 will become

equation_1g1 = " = a*s^2+d"

reflectance_red_g1= red_values[0]*solar_zenith_rad_squared+red_values[3]
reflectance_nir_g1= nir_values[0]*solar_zenith_rad_squared+nir_values[3]
```

# G-2 Conditions Calculations for Red and NIR

```python
#Assuming Condition G-2 i.e. Western Swath Center where v = 7.5 and v-s=0
equation_1= " = a(v^2+s^2)+b(v^2s^2)+cvsCos(v-s)+d"
#as v-s=0 so cos (0)=1. substituting this in equation-1 
#also putting values of v= 7.5 in equation 1
equation_1_g2= " = a(7.5^2+s^2)+b(7.5^2*s^2)+c7.5*s+d"

view_zenith= (7.5*(np.pi/180))

reflectance_red_g2= red_values[0]*(np.square(view_zenith)+solar_zenith_rad_squared)+red_values[1]*(np.square(view_zenith)+solar_zenith_rad_squared)+red_values[2]*(view_zenith*solar_zenith_rad)+red_values[3]
reflectance_nir_g2= nir_values[0]*(np.square(view_zenith)+solar_zenith_rad_squared)+nir_values[1]*(np.square(view_zenith)+solar_zenith_rad_squared)+nir_values[2]*(view_zenith*solar_zenith_rad)+nir_values[3]
```

# G-3 Conditions Calculations for Red and NIR

```python
#Assuming Condition G-3 i.e. Eastern Swath Center where v = 7.5 and v-s=0
equation_1= " = a(v^2+s^2)+b(v^2s^2)+cvsCos(v-s)+d"
#as v-s=180 so cos (180)=-1. substituting this in equation-1 
#also putting values of v= 7.5 in equation 1
#This will only reverse the sign of term which includes c and the remaining will be the same
equation_1_g2= " = a(7.5^2+s^2)+b(7.5^2*s^2)-cvs+d"

view_zenith= (7.5*(np.pi/180))

reflectance_red_g3= red_values[0]*(np.square(view_zenith)+solar_zenith_rad_squared)+red_values[1]*(np.square(view_zenith)+solar_zenith_rad_squared)-red_values[2]*(view_zenith*solar_zenith_rad)+red_values[3]
reflectance_nir_g3= nir_values[0]*(np.square(view_zenith)+solar_zenith_rad_squared)+nir_values[1]*(np.square(view_zenith)+solar_zenith_rad_squared)-nir_values[2]*(view_zenith*solar_zenith_rad)+nir_values[3]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,8)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(months, reflectance_red_g1, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Swath Center')
ax1.plot(months, reflectance_red_g2, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Western Edge')
ax1.plot(months, reflectance_red_g3, color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Eastern Edge')
ax1.legend()
ax1.set(xlabel='Mid-month i.e 15th of each month', ylabel='Value',
title='Red Reflectance under Different Geometric Conditions')
ax1.grid()
ax1.set_ylim([0, 0.7])

ax2.plot(months, reflectance_nir_g1, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Swath Center')
ax2.plot(months, reflectance_nir_g2, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Western Edge')
ax2.plot(months, reflectance_nir_g3, color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Eastern Edge')
ax2.set(xlabel='Mid-month i.e 15th of each month', ylabel='Value',
title='NIR Reflectance under Different Geometric Conditions')
ax2.grid()
ax2.set_ylim([0, 0.7])
ax2.legend()



fig.savefig("A3task4.png", dpi=600)
plt.show()
```
