```python
#This is the first assignment of Quantitative Remote Sensing for Terrestrial Monitoring (GEOG/GSE-741-S01)
#Import necessary libraries for processing
import os 
import matplotlib as mtp
import rasterio as rio
from rasterio.plot import show
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt

spectral_WL=np.array(["0.450-0.515", "0.525-0.600", "0.630-0.680", "0.845-0.885","1.560-1.660","2.100-2.300"])
print("Libraries Imported Successfully")

print(os.getcwd())
one_folder_up = os.path.dirname(os.getcwd())
data_dir= os.path.join(one_folder_up,"data_a2")
```

```python
data_url=[] #for ease of access all the paths of TIFF files will be stored in this array so that it is easy later on in this program
dataset=[] #this will be used for just reading each file and put the object into array
#print(data_dir)
ext=['.TIF', 'tif']
for filename in os.listdir(data_dir):
    if filename.endswith(tuple(ext)): 
        filepath= os.path.join(data_dir, filename)
        data_url.append(filepath)

#Make a list of unccessary files and create another list to store their paths

file_to_remove= ['T1_SR_B1.TIF','T1_ST_B10.TIF','T1_QA_PIXEL.TIF','T1_QA_RADSAT.TIF','T1_SR_QA_AEROSOL.TIF','T1_ST_ATRAN.TIF','T1_ST_CDIST.TIF','T1_ST_DRAD.TIF','T1_ST_EMIS.TIF','T1_ST_EMSD.TIF','T1_ST_QA.TIF','T1_ST_TRAD.TIF','T1_ST_URAD.TIF']
file_to_remove_path= []
for file_name in data_url:
    if file_name.endswith(tuple(file_to_remove)):
        file_to_remove_path.append(file_name)

#Remove all those files from data_url list so that we are left only with the necessary bands
for a in file_to_remove_path:
    #print(a)
    data_url.remove(a)
```

```python
for band_url in data_url:
    #dataset.append(rio.open(band_url))#this will open each datatset in read mode
    with rio.open(band_url) as f:
        dataset.append(f.read(1))
    #just to make sure paths are working
```

```python
landsat_img = np.array(dataset, dtype=dataset[0].dtype)
# let's check the shape of this array
landsat_img.shape
```

```python
#above results tell us we have 6 bands each having number of rows and columns as shown
#show(sentinel_img[0,1,2])
#show(sentinel_img[[3], :, :]) #NIR Band
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
image = landsat_img[2, :, :] 
plt.imshow(image,interpolation='nearest')
plt.imsave('Figur1_wholeimage.png', image)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15,10))
clipped_img = landsat_img[:, 6500:6900, 4100:4300]
print(clipped_img.shape)
plt.imshow(clipped_img[3, :, :],interpolation='nearest')
#plt.show(clipped_img[[5,4], :, :])
plt.imsave('Figur2_Subset.png', clipped_img[3,:,:])
```

```python
#going closer a little more so that we can get accurate
plt.figure(figsize = (15,10))
clipped_img = landsat_img[:, 6720:6800, 4125:4245]
#print(clipped_img.shape)
plt.imshow(clipped_img[3, :, :],interpolation='nearest',cmap="RdYlGn")
plt.colorbar()
plt.imsave('Figur3_Subset.png', clipped_img[3,:,:], dpi=600)
#plt.show(clipped_img[[5,4], :, :])

print ("Cell Selected for Water is Row 30 and Column 80. DN is: ",clipped_img[3][30][80])
print ("Cell Selected for Vegtatation is  is Row 10 and Column 3. DN is: ",clipped_img[3][10][3])
print("Above values were selected based on observation from NIR band of imagery and Vegtation Index")
```

```python
#Now the next step is extract values for these cells from the original image from each band
water= clipped_img[:, 30, 80]
vegetation= clipped_img[:, 10, 3]
print("value of water in all selected bands i.e. 2-7: ", water)
print("value of vegetation in all selected bands i.e. 2-7: ", vegetation)
```

```python
#Conversion of Reflective Band pixels values to surface reflectance
#declaring variables
# l=m+a where l is TOA spectral radiance, m is band specific multiplicative rescaling factor and a is band specific additive rescaling factor


radiance_mult_fac= 2.75e-05 #[1.2652E-02,1.1658E-02,9.8310E-03,6.0161E-03,1.4961E-03,5.0428E-04]
additive_resc_fac= -0.2 #[-63.25803,-58.29174,-49.15488,-30.08033,-7.48071,-2.52140]

l_water = water*radiance_mult_fac+additive_resc_fac
l_veg = vegetation*radiance_mult_fac+additive_resc_fac

#above two variables stores TOA spectral randiance for all desired bands for two pixels i.e. water and vegetation as shown below
print("Surface Reflectance values for Water", l_water)
print("Surface Reflectance values for Vegetation", l_veg)
```

```python
toa_water = [0.12663797, 0.11552773, 0.11581944, 0.11047188, 0.04503562, 0.01736783]
toa_vege = [0.11725684, 0.11050556, 0.09049771, 0.32422748, 0.14908407, 0.0426173 ]

```

## Task:Pixels Surface Reflectance vs TOA reflectance

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,5)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(spectral_WL, l_water, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='SR Water')
ax1.plot(spectral_WL, toa_water, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='TOA Water')
ax1.legend()
ax1.set(xlabel='Wavelength μm', ylabel='Value',
title='TOA vs Surface Reflectance of Water Pixel')
ax1.grid()
ax1.set_ylim([0, 0.30])



ax2.plot(spectral_WL, l_veg, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='SR Vegetation')
ax2.plot(spectral_WL, toa_vege, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='TOA Vegetation')
ax2.set(xlabel='Wavelength μm', ylabel='Value',
title='TOA vs Surface Reflectance of Vegetation Pixel')
ax2.grid()
ax2.legend()

fig.savefig("task2-1.png")
plt.show()
```

```python
difference_water= toa_water-l_water
difference_veg= toa_vege-l_veg
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(18,5)) # 2 axes on a 1x2 grid

#fig, ax = plt.subplots()
ax1.plot(spectral_WL, difference_water, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Difference Water')
#ax1.plot(spectral_WL, toa_water, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='TOA Water')
ax1.legend()
ax1.set(xlabel='Wavelength μm', ylabel='Value',
title='Difference in TOA and  surface for water pixel')
ax1.grid()



ax2.plot(spectral_WL, difference_veg, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='Difference Vegetation')
#ax2.plot(spectral_WL, toa_vege, color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12, label='TOA Vegetation')
ax2.set(xlabel='Wavelength μm', ylabel='Value',
title='Difference in TOA and  surface for vegetation Pixel')
ax2.grid()
ax2.legend()

fig.savefig("differencesTOA&SR.png")
plt.show()
```
